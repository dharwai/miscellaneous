```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'CK RESNET.csv'  # Replace with your CSV file path
data = pd.read_csv(file_path)

# Plotting Loss and Validation Loss
plt.figure(figsize=(12, 6))

# Plot 1: Loss
plt.subplot(1, 2, 1)
plt.plot(data['Loss'], label='Training Loss', color='blue')
plt.plot(data['Val_Loss'], label='Validation Loss', color='orange')
plt.title('Loss') 
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot 2: Accuracy
plt.subplot(1, 2, 2)
plt.plot(data['Accuracy'], label='Training Accuracy', color='blue')
plt.plot(data['Val_Accuracy'], label='Validation Accuracy', color='orange')
plt.title('Accuracy ')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
```
