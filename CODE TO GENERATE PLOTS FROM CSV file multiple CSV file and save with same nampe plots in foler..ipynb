```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'ck+vgg1.csv'  # Replace with your CSV file path
data = pd.read_csv(file_path)

# Plotting Loss and Validation Loss
plt.figure(figsize=(12, 6))

# Plot 1: Loss
plt.subplot(1, 2, 1)
plt.plot(data['Loss'], label='Training Loss', color='blue')
plt.plot(data['Val_Loss'], label='Validation Loss', color='orange')
plt.title('Loss') 
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot 2: Accuracy
plt.subplot(1, 2, 2)
plt.plot(data['Accuracy'], label='Training Accuracy', color='blue')
plt.plot(data['Val_Accuracy'], label='Validation Accuracy', color='orange')
plt.title('Accuracy ')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()
    
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Folder path containing CSV files
folder_path = 'C:/Users/RAJESH KUMAR/Desktop/avalues extravtion/NEW_EXP_Values/EXP6'  # Replace with your folder path

# Iterate through each CSV file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        # Load the CSV file
        file_path = os.path.join(folder_path, file_name)
        data = pd.read_csv(file_path)

        # Plotting Loss and Validation Loss
        plt.figure(figsize=(12, 6))

        # Plot 1: Loss
        plt.subplot(1, 2, 1)
        plt.plot(data['Loss'], label='Training Loss', color='blue')
        plt.plot(data['Val_Loss'], label='Validation Loss', color='orange')
        plt.title('Loss')
        plt.xlabel('Epochs')
        plt.ylabel('Loss')
        plt.legend()

        # Plot 2: Accuracy
        plt.subplot(1, 2, 2)
        plt.plot(data['Accuracy'], label='Training Accuracy', color='blue')
        plt.plot(data['Val_Accuracy'], label='Validation Accuracy', color='orange')
        plt.title('Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()

        # Save the plot with the same name as the CSV file (but with .png extension)
        output_file = os.path.join(folder_path, f"{os.path.splitext(file_name)[0]}.png")
        plt.tight_layout()
        plt.savefig(output_file, dpi=300)
        plt.close()

        print(f"Plot saved for {file_name} as {output_file}")
```
